```python
def lookup_lemmas(filename):
    chunk_num = 0
    chunks = {} # map chunk number to list of records. a record is (pos, lemma, stem, index)
    s = {}
    records = []
    with open(filename, encoding="utf-8-sig", newline='') as f:
        for line in f:
            if line[:6] == "CHUNK:":
                s = {}
                chunks[chunk_num] = records
                chunk_num += 1
                records = []
            if line[:2] == 's:':
                s_line_n = int(line.split('·')[0][2:])
                snum1 = line.split('·')[3]
                snum2 = line.split('·')[4]
                s_t_no1 = line.split('·')[5]
                s_t_no2 = line.split('·')[6]
                stem = line.split('·')[2]
                lemma = line.split('·')[8]
                
                s[(s_t_no1, s_t_no2)] = {"s_n":s_line_n, 
                                         "stem": line.split('·')[2], 
                                          "lemma": line.split('·')[8],
                                            'snum1': line.split('·')[3],
                                            'snum2': line.split('·')[4]}
                
            if line[:2] == 't:': 
                pos = line.split('·')[1]
                if 'NOUN' in pos or 'ADJ' in pos:
                    t_line_n = int(line.split('·')[0][2:])
                    tnum1 = line.split('·')[6]
                    tnum2 = line.split('·')[7] 
                    
                    for keys, values in s.items():
                        if int(t_line_n) >= int(keys[0]) and int(t_line_n) <= int(keys[1]):
                            sn = str(s[keys]['s_n'])
                            lemma = s[keys]['lemma']   
                            stem = s[keys]['stem']
                            num1 = s[keys]['snum1']
                            num2 = s[keys]['snum2']
                            record = {"s_n": sn, 's_keys': keys, 't_n': t_line_n, 'pos': pos, 'lemma': lemma, 
                                      'stem': stem, 'index': (num1, num2)}
                            records.append(record)  
    #print(len(records))
    #print(records)
    chunks[chunk_num] = records
    del chunks[0]
    return chunks

#chunks = lookup_lemmas('C:/Users/19061/Desktop/Arabic NLP/data/integrated/20000715_AFP_ARB.0001.txt')
#print(chunks)
#items = []
#for chunk in chunks.values():
    #for record in chunk:
        #items.append(record['lemma'])
#countss = Counter(items)
#print(sum(counts.values()))
#print(countss)
```

```python
import os
#path = 'C:/Users/19061/Desktop/Arabic NLP/data/integrated/'
path = 'C:/Users/19061/Desktop/integrated_all/'

def lookup_all_lemmas(path):
    files = os.listdir(path)
    results = {}
    for file in files:
        results[file] = lookup_lemmas(path + file)
    return results

all_data = lookup_all_lemmas(path)
```
