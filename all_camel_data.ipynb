```python
lexemes = dict()
with open('C:/Users/19061/Downloads/morphology.db.bw') as f:
    k = f.readlines()
    k = k[1364:74100]
    c = 0
    print(len(k))
    for i in k:
        if( len(i.split("\t")) <2):
            print(i)

        item0, item1, item = i.split("\t", 2)
        item = item.split(" ")
        lexeme = []
        record = dict()
        pos = []
        num ="s"
        form_num = "s"
        for j in item:
            #print(j)
            key, value = j.split(':',1)
            if key in ['diac', 'root', 'pattern', 'gloss']:
                record.update({key:value.strip()})
            elif key == 'lex':
                lexeme = value.strip()
            elif key == 'num':
                num = value.strip()
            elif key == 'form_num':
                form_num = value.strip()
            elif key == 'pos':
                pos = value.strip()
            if key in ['num', 'form_num', 'pos']:
                record.update({key:value.strip()})
        if(  pos == 'noun' or pos== 'adj'): ## anything else??
            record.update({'stem_class':item1})
            record.update({'broken': False})
            if lexeme.strip() == 'muEal~im_1' and num == 'p': ##checking if it exists?
                print(form_num)
        
            if num == '-':
                if form_num == '-':
                    lexeme += '_-'
                elif form_num == "s":
                    #print("here")
                    lexeme += '_s'
                elif form_num == 'p': ##no lexemes enter this        
                    lexeme += '_p'
                

            else:
                lexeme += ("_"+num)
                #print(num)
                if(form_num == 's' and num == 'p'):
                    record.update({'broken': True})

            if lexeme in lexemes.keys():
                record_list = lexemes[lexeme]
                record_list.append(record)
                lexemes.update({lexeme:record_list})
            else:
                record_list = list()
                record_list.append(record)
                lexemes.update({lexeme: record_list})
```

```python
sp = []
for i in bw_ls:
    s = i + '_s'
    p = i + '_p'
    sp.append(s)
    sp.append(p)
lexs = []
for i in sp:
    if i in list(lexemes.keys()):
        lexs.append(i)
print(len(lexs))
print(lexs)  # 12222 ~  15242
```

```python
IC = dict()
count = 0
for lex in lexs:
    if(lex[-1] == "p"):
        lex_p = lex
        lex_s = lex[0:-1] + "s"
        #print(lex_s, " ", lex_p)
        if lex_s in lexs:
            for i in range(len(lexemes[lex_s])):
                for j in range(len(lexemes[lex_p])):
                    if(lexemes[lex_p][j]['broken'] == True):
                        ic_key = lexemes[lex_s][i]['pattern'] + ' , ' +lexemes[lex_p][j]['pattern']  
                        ic_value = dict() #lemma, root, gloss
                        ic_value.update({'lemma': lex_s[0:-2]})
                        ic_value.update({'diac_sing': lexemes[lex_s][i]['diac']})
                        ic_value.update({'diac_plu': lexemes[lex_p][j]['diac']})
                        ic_value.update({'stem_class_sing': lexemes[lex_s][i]['stem_class']})
                        ic_value.update({'stem_class_plu': lexemes[lex_p][j]['stem_class']})
                        ic_value.update({'root': lexemes[lex_s][i]['root']})
                        ic_value.update({'gloss': lexemes[lex_s][i]['gloss']})
                        if 'ic_classes' not in lexemes[lex_s][i]:
                            lexemes[lex_s][i].update({'ic_classes':list()})
                        if ic_key not in lexemes[lex_s][i]['ic_classes']:
                            lexemes[lex_s][i]['ic_classes'].append(ic_key)
                        if ic_key not in IC:
                            IC.update({ic_key: list()})
                        count+=1
                        IC[ic_key].append(ic_value)
print(len(IC.keys()))    #1266  ~  1970
print(IC)
```

```python
d = get_frq_bw_words()
l = {}
for keys, values in IC.items():
    words = []
    for value in values:
        words.append(value['lemma'])
    f = 0
    for word in words:
        f += int(d[word])
    sum = f*len(words)
    print(keys, ': ', len(IC[keys]), ' ', words, ' ', sum)
    l[keys] = sum
```

```python
# top 200 IC classes (token counts)
import operator
sorted_l = dict( sorted(l.items(), key=operator.itemgetter(1),reverse=True))
#print('Dictionary in descending order by value : ',sorted_l)
#dict(sorted(l.items(), key=lambda item: item[1]))
top_ICs = []
for keys, values in sorted_l.items():
    top_ICs.append(keys)
print(len(top_ICs))
print(top_ICs[0:200])
```

```python
# 100, 200 lemma (how many IC are represented in the first 100 words)  hopefully lower IC classes (regular 10 IC classes)
ls_ic = []
ls_top_ic = []
for i in lexs[0:12222]:
    #print(lexemes[i][0])
    for keys, values in lexemes[i][0].items():
        if keys == 'ic_classes':
            for v in values:
                ls_ic.append(v)
                if v in top_ICs[0:200]:
                    ls_top_ic.append(v)
print("no. of IC classes:  ", len(set(ls_ic)))  
print("no. of top IC classes:  ", len(set(ls_top_ic))) 

# 100 -> 39,38,   200->72, 65, 500->134, 90, 1000->218, 132, 5000->504, 163, 10000 -> 753, 168, 12222->843, 168


# a lot ICs are rare, after 1000th (not available in language learning);
# language learner's persepetive: how many IC classes from 100 lemmas
```
